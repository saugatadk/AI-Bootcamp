## Data Cleaning and Handling Missing Values

```python
# Check for missing values
missing_values = music.isnull().sum()
print("\nMissing values in each column:")
missing_values
```

```python
# Handling Missing Values

music_copy = music.copy() 

# Drop rows with missing values
print(f"Original shape: {music_copy.shape}")
complete_music = music_copy.dropna()
print(f"After dropping rows with NaN: {complete_music.shape}")
```

```python
# Drop specific columns with missing values
music_drop_cols = music_copy.dropna(axis=1)
print(f"After dropping columns with NaN: {music_drop_cols.shape}")
```

```python
if 'tags' in music_copy.columns:
    music_copy.fillna({'tags': 'Not Found'}, inplace = True)
    print("Filled missing tags with 'Not Found'")

# Verify the filling worked
print(f"\nMissing values after filling: {music_copy.isnull().sum()}")
```

# Data Filtering and Querying

```python
# Single Condition
billion_viewed_music = music_copy[music_copy['view_count'] >= 1000000000]
print("Count of music viewed billion times in 2025: ", len(billion_viewed_music))
billion_viewed_music
```

```python
# Multiple condition 
billion_viewed_music_LadyGaga = music_copy[(music_copy['view_count'] >= 1000000000) & (music_copy['channel'] == 'Lady Gaga')]
billion_viewed_music_LadyGaga
```

```python
# Using isin()
eilish_gaga_music = music_copy[music_copy['channel'].isin(['Billie Eilish', 'Sean Paul'])]
eilish_gaga_music
```

```python
# Using query()
query = music_copy.query('channel_follower_count > 1000000 and view_count > 1000000')
print("Number of music with more than 1 million views and more than 1 million followers: ", len(query))
```

# Data Aggregation and Grouping

```python
# Basic Aggregation Statistics
if 'duration' in music_copy.columns:
    print(f"Mean duration: {music_copy['duration'].mean():.2f}s")
    print(f"Median duration: {music_copy['duration'].median():.2f}s")
    print(f"Standard deviation of duration: {music_copy['duration'].std():.2f}s")
```

```python
# Value counts
if 'channel' in music_copy.columns:
    artist = music_copy['channel'].value_counts()
    print("Number of music tracks by artist:\n", artist)
```

```python
# Average duration of music by views count by applying categorization
if 'duration' in music_copy.columns and 'view_count' in music_copy.columns:
    # Create view count categories
    def categorize_views(views):
        if views < 100_000:
            return 'Under 100K'
        elif views < 500_000:
            return '100K - 500K'
        elif views < 1_000_000:
            return '500K - 1M'
        elif views < 1_000_000_000:
            return '1M - 1B'
        else:
            return 'Over 1B'
    # Apply categorization
    music_copy['view_category'] = music_copy['view_count'].apply(categorize_views)
    duration_by_views = music_copy.groupby('view_category')['duration'].mean()
    print("Average Duration by Views")
    print(duration_by_views)
    
    # Show count in each category
    print("\nNumber of songs in each category:")
    print(music_copy['view_category'].value_counts())
    
```

# Pivot Table and Data Reshaping

```python
# Pivot table
if 'duration' in music_copy.columns and 'channel' in music_copy.columns and 'view_category' in music_copy.columns:
    pivot_duration = pd.pivot_table(
        music_copy,
        values='duration',
        index='channel',
        columns='view_category',
        aggfunc='mean',
        fill_value=0
    )
    print("Average Duration by Channel and View Category:")
    print(pivot_duration.head())
```

```python
# Data Reshaping: Melt Example
# Create a sample subset for melting
music_subset = music_copy[['title', 'channel', 'view_count', 'duration', 'channel_follower_count']].head(10)

# Melt the data to long format
melted_music = pd.melt(
    music_subset,
    id_vars=['title', 'channel'],
    value_vars=['view_count', 'duration', 'channel_follower_count'],
    var_name='metric',
    value_name='value'
)
print("Melted Data (Long Format):")
melted_music.sample(10)
```
